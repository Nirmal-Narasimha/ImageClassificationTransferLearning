#### Imports

```python
import tensorflow as tf
from tensorflow.keras.utils import image_dataset_from_directory
from tensorflow.keras.models import Model,Sequential
from tensorflow.keras.regularizers import l2
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint,ReduceLROnPlateau
from tensorflow.keras.layers import RandomCrop,RandomFlip, RandomRotation, RandomZoom, RandomContrast, RandomTranslation,Dense, Dropout, BatchNormalization, GlobalAveragePooling2D
from tensorflow.keras.applications import ResNet50, ResNet101, EfficientNetB0, VGG16
from tensorflow.keras.applications.efficientnet import preprocess_input as efficientnet_preprocess
from tensorflow.keras.applications.vgg16 import preprocess_input as vgg16_preprocess
from tensorflow.keras.applications.resnet50 import preprocess_input as resnet50_preprocess
from tensorflow.keras.applications.resnet import preprocess_input as resnet101_preprocess
import matplotlib.pyplot as plt
import os
from sklearn.metrics import classification_report, roc_auc_score
import numpy as np
from sklearn.model_selection import train_test_split
import warnings
warnings.filterwarnings("ignore")
```

#### 1(b). Data Loading Exploration and Preprocessing

```python
#ChatGPT Prompt- help me remove .DSStore file thats showing up in my train directory
def removeDsstore(directory):
    for root, dirs, files in os.walk(directory):
        for file in files:
            if file == '.DS_Store':
                os.remove(os.path.join(root, file))
removeDsstore('../data/seg_train')
removeDsstore('../data/seg_test')
```

```python
trainDir = '../data/seg_train'
testDir = '../data/seg_test'
```

```python
#Get train data image paths and class labels
#https://www.geeksforgeeks.org/python-os-listdir-method/
#https://docs.python.org/3/library/os.path.html
def getFilePathAndLabels(dir):
  filePaths = []
  labels = []
  classes = sorted(os.listdir(dir))
  for label,className in enumerate(classes):
    classDir = os.path.join(dir,className)
    for img in os.listdir(classDir):
      filePaths.append(os.path.join(classDir,img))
      labels.append(label)
  return np.array(filePaths),np.array(labels),classes
filePaths,labels,classNames = getFilePathAndLabels(trainDir)
```

```python
#Hyperparameters
#Data Params
#https://piazza.com/class/m09cnglrrao1gr/post/1156
bSize = 64
validSize = 0.2
randomS = 42
imgSize = (224,224)
noClasses = len(classNames) 
#model Params
epochNo = 50
modelIpShape = (224,224,3)
l2Reg = 0.01
dropFactor = 0.2
lr = 0.001
decayStep=1000
decayRate=0.9
earlyStopPatience =20
```

```python
#Stratify the train set to train and validation set selecting 20% from each class in validation set
#https://stackoverflow.com/questions/29438265/stratified-train-test-split-in-scikit-learn
trainPaths,valPaths,trainLabels,valLabels= train_test_split(filePaths,labels,test_size=validSize,stratify=labels,random_state=randomS)
```

#### 1(b).ii

```python
#Create Tf Datasets and preprocess image for model specific requirements from tensorflow preprocess function
#https://stackoverflow.com/questions/49579684/what-is-the-difference-between-dataset-from-tensors-and-dataset-from-tensor-slic
#https://www.tensorflow.org/api_docs/python/tf/image
def createTfDataset(filePaths, labels, batchSize, imageSize,preprocess_fn=None, shuffle=False):
    dataset = tf.data.Dataset.from_tensor_slices((filePaths, labels))
    if shuffle:
        dataset = dataset.shuffle(buffer_size=len(filePaths), reshuffle_each_iteration=True)
    def preprocess(filePath, label):
        img = tf.io.read_file(filePath)
        img = tf.image.decode_jpeg(img, channels=3)
        img = tf.image.resize(img, imageSize)
        if preprocess_fn:
            img = preprocess_fn(img) 
        return img, tf.one_hot(label, depth=len(classNames))
    return dataset.map(preprocess).batch(batchSize).prefetch(tf.data.AUTOTUNE)
```

#### 1(c).ii Data Augmentation - Empirical Regularization

```python
#Function to augment the train set
#https://www.tensorflow.org/tutorials/images/data_augmentation
dataAug = tf.keras.Sequential([    
     RandomFlip("horizontal_and_vertical"),
     RandomRotation(0.2),
     RandomZoom(0.1),
     RandomContrast(0.1),
     RandomTranslation(0.1,0.1),
     tf.keras.layers.Resizing(224, 224)
])
```

```python
trainData = createTfDataset(trainPaths, trainLabels, batchSize=bSize, imageSize=imgSize, shuffle=True)
valData = createTfDataset(valPaths, valLabels, batchSize=bSize, imageSize=imgSize, shuffle=False)
augTrainData = trainData.map(lambda x, y: (dataAug(x, training=True), y)).prefetch(buffer_size=tf.data.AUTOTUNE)
```

```python
#Preprocessing the test data-No Augmentation
#https://www.tensorflow.org/api_docs/python/tf/keras/preprocessing/image_dataset_from_directory
def preprocessTestImgData(image, label):
    image = tf.cast(image, tf.float32)
    label = tf.one_hot(label, depth=len(classNames))
    return image,label

testData = image_dataset_from_directory(
    testDir,
    image_size=imgSize, 
    batch_size=bSize
).map(preprocessTestImgData).prefetch(buffer_size=tf.data.AUTOTUNE)
```

#### Model Training and Evaluation

```python
#Helper to plot learning Curves
#https://matplotlib.org/3.5.3/api/_as_gen/matplotlib.pyplot.html
#https://stackoverflow.com/questions/37664783/how-to-plot-a-learning-curve-for-a-keras-experiment
def plotTrainCurve(history,baseModelName):
    plt.figure(figsize=(12,5))
    #Loss Curve
    plt.subplot(1,2,1)
    plt.plot(history.history["loss"],label="Training Loss")
    plt.plot(history.history["val_loss"],label="Validation Loss")
    plt.title(f"{baseModelName} Loss Curve")
    plt.xlabel("Epochs")
    plt.ylabel("Loss")
    plt.legend()
    #Accuracy
    plt.subplot(1,2,2)
    plt.plot(history.history["accuracy"],label="Training Accuracy")
    plt.plot(history.history["val_accuracy"],label="Validation Accuracy")
    plt.title(f"{baseModelName} Accuracy Curve")
    plt.xlabel("Epochs")
    plt.ylabel("Accuracy")
    plt.legend()
    plt.show()
```

```python
#Helper to get Model Performance Metrics
#https://towardsdatascience.com/multiclass-classification-evaluation-with-roc-curves-and-roc-auc-294fd4617e3a
#https://scikit-learn.org/1.5/modules/generated/sklearn.metrics.classification_report.html
def getMetrics(model,trainData,valData,testData,baseModelName,numClasses=noClasses):
  def ComputeMetrics(data,datasetName):
    YTrue = []
    YPred = []
    YPredProbs = []
    for img,label in data:
      YTrue.extend(tf.argmax(label,axis=1).numpy())
      preds = model.predict(img,verbose=0)
      YPred.extend(tf.argmax(preds,axis=1).numpy())
      YPredProbs.extend(preds)
    YPredProbs = np.array(YPredProbs)
    print(f"Full Classification Report for {baseModelName} on {datasetName} data:")
    print(classification_report(YTrue, YPred,target_names=classNames))
    #AUC
    YTrueOneHot = tf.one_hot(YTrue,depth=numClasses).numpy()
    auc = roc_auc_score(YTrueOneHot,YPredProbs,multi_class="ovr")
    print(f"AUC for {baseModelName} on {datasetName} data:{auc}")

  print("Metrics for Train Data:")
  ComputeMetrics(trainData,"Training")
  print("Metrics for Validation Data:")
  ComputeMetrics(valData,"Validation")
  print("Metrics for Test Data:")
  ComputeMetrics(testData,"Test")
```

```python
#Training the model
#https://www.tensorflow.org/api_docs/python/tf/keras/optimizers/schedules/ExponentialDecay
#https://www.tensorflow.org/api_docs/python/tf/keras/callbacks/EarlyStopping
#https://www.tensorflow.org/api_docs/python/tf/keras/callbacks/ModelCheckpoint
#https://www.geeksforgeeks.org/adam-optimizer-in-tensorflow/
#https://keras.io/api/layers/pooling_layers/global_average_pooling2d/
#https://www.kaggle.com/code/ryanholbrook/dropout-and-batch-normalization
#https://www.kaggle.com/discussions/questions-and-answers/95792
def modelBuild(baseModelName,baseModelFn,trainData,valData,testData,noClasses=noClasses,epochs=epochNo,batchSize=bSize):
  baseModel = baseModelFn(weights="imagenet",include_top=False,input_shape=modelIpShape)#remove ultimate layer
  baseModel.trainable = False #Freeze the model

  model = Sequential([
    baseModel,
    GlobalAveragePooling2D(),
    Dense(256, activation="relu", kernel_regularizer=l2(l2Reg)),
    BatchNormalization(),
    Dropout(dropFactor),
    Dense(noClasses, activation="softmax")  # Output layer
])
  #Learning rate schedule to decay lr over time
  LRscheduler = tf.keras.optimizers.schedules.ExponentialDecay(
    initial_learning_rate=lr,
    decay_steps=decayStep,
    decay_rate=decayRate
)
  model.compile(
      optimizer = tf.keras.optimizers.Adam(learning_rate=LRscheduler),
      loss = "categorical_crossentropy",
      metrics = ["accuracy"]
  )
  
  checkPointPath = f'checkpoints/{baseModelName}_best.keras'
  callbacks = [
      ModelCheckpoint(
            filepath=checkPointPath,  # Save checkpoint
            save_best_only=True,      # Only save the best model
            monitor="val_loss",       # Monitor validation loss
            mode="min",               # Save when 'val_loss' decreases
            verbose=1
        ),
    EarlyStopping(
        monitor="val_loss",
        patience=earlyStopPatience,  # Wait for improvement for up to 20 epochs
        restore_best_weights=True
    )
]

  #Train
  history = model.fit(
      trainData,
      validation_data = valData,
      epochs = epochs,
      callbacks = callbacks
  )
  if os.path.exists(checkPointPath):
        print(f"Loading the best weights for {baseModelName} from checkpoint...")
        model.load_weights(checkPointPath)
  else:
    print(f"Checkpoint not found at {checkPointPath}. Skipping weight loading.")

  #Eval on Test Set
  print(f"Evaluating model:{baseModelName}")
  testLoss,testAccuracy = model.evaluate(testData)
  print(f"Test Loss:{testLoss}\nTest Accuracy:{testAccuracy}")
  plotTrainCurve(history,baseModelName)
  getMetrics(model,trainData,valData,testData,baseModelName,noClasses)
  model.save(f"{baseModelName}.keras")
  return model
```

```python
# Train and evaluate ResNet50
#https://keras.io/api/applications/resnet/
resNet50_model = modelBuild(
    baseModelName="ResNet50",
    baseModelFn=ResNet50,
    trainData=augTrainData,
    valData=valData,
    testData=testData,
    noClasses=noClasses,
    epochs=epochNo,
    batchSize=bSize
)
```
